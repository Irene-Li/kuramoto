```python
import numpy as np
from matplotlib import pyplot as plt 
```

```python
def U(phi, eta): 
    return eta*(phi**3/3 - phi)
```

```python
phi = np.linspace(-2.5, 2.5, 100)
U1 = U(phi, 1)
U2 = U(phi, -1)
plt.rc('text', usetex=True)
plt.rc('font', size=20, family='serif')

fig, axes = plt.subplots(1, 2, sharex=True, sharey=True, figsize=(9.5, 5))
axes[0].axvline(x=-1, color='orange', linestyle='--')
axes[0].axvline(x=1, color='orange', linestyle='--')
axes[0].plot(phi, U1, 'k-')
axes[0].set_xlim([-2.5, 2.5])
axes[0].set_xticks([-1, 0, 1])
axes[0].set_xticklabels([r'$-\sqrt{\Omega/\eta}$', '0', r'$\sqrt{\Omega/\eta}$'])
axes[0].set_xlabel(r'$\phi$')
axes[0].set_yticks([-2, 0, 2])
axes[0].set_ylabel(r'$U(\phi)$')
axes[0].set_title(r'Positive $\eta$')
axes[1].plot(phi, U2, 'k-')
axes[1].set_xlabel(r'$\phi$')
axes[1].set_title(r'Negative $\eta$')
axes[1].axvline(x=-1, color='orange', linestyle='--')
axes[1].axvline(x=1, color='orange', linestyle='--')
plt.tight_layout()
plt.savefig('potential.pdf')
plt.show() 
```
